```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
sah=pd.read_csv('energy.csv')
print(sah)


srh=sah[sah['State']=="Andhra"]
rcb=sah[sah["State"]=='Karnataka']

fig1,diag1=plt.subplots(2)
diag1[0].plot(srh['Date'],srh['Consumption'])
diag1[1].plot(rcb['Date'],rcb['Consumption'])


#barchart telengana
revanth=sah[sah['State']=='Telengana']
cm=sah[sah["State"]=='Tripura']

fig,diag=plt.subplots(2)
diag[0].bar(revanth['Date'],revanth['Consumption'],color='green')
diag[1].bar(cm['Date'],cm['Consumption'],color='blue')


#pie chart
state1='Delhi'
state2='Odisha'
state3='Up'
state4='Punjab'
cse5=[state1,state2]
cse4=sah[sah['State']==state1]['Consumption'].sum()
cse3=sah[sah['State']==state2]['Consumption'].sum()
cse2=[cse4,cse3]
fig,diag=plt.subplots(2)
diag[0].pie(cse2,labels=cse5,autopct='1%.1f%%')


cse5=[state3,state4]
cse4=sah[sah['State']==state3]['Consumption'].sum()
cse3=sah[sah['State']==state4]['Consumption'].sum()
cse2=[cse4,cse3]
diag[1].pie(cse2,labels=cse5,autopct='1%.1f%%')
```
